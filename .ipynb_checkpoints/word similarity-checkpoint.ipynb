```python
#resume modeling:
def resume_reader(example, keyword):
    toke_example = pda.tokenize_stem(example)
    example_words = toke_example[0].split()
    matching_words = []
    missing_words = []
    for word in keywords[keyword]:
        if word in example_words:
            matching_words.append(word)
        else:
            missing_words.append(word)
    return matching_words, set(missing_words)
```

```python
resumes = pd.read_csv('resumes.csv')
example = pd.Series(resumes.Resumes[0])
```

```python
resume_reader(example, 'marketing')
```
